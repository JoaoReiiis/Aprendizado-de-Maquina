```python
import spacy
import json

nlp = spacy.load("pt_core_news_lg")
with open("dados/resultados.json", "r", encoding="utf-8") as f:
    lista_de_dados = json.load(f)

for item in lista_de_dados:
        texto_para_analise = item["conteudo_texto"]
        print(f"Processando texto: {texto_para_analise[:50]}...")
        doc = nlp(texto_para_analise)
        print("Categorias de palavras (Entidades) encontradas:")
        if not doc.ents:
            print("Nenhuma entidade foi reconhecida pelo modelo large neste texto.")
        else:
            for entidade in doc.ents:
                print(f"- Texto da Entidade: '{entidade.text}'")
                print(f"  Categoria (Label): {entidade.label_}")
                print(f"  Explicação: {spacy.explain(entidade.label_)}\n")
```
