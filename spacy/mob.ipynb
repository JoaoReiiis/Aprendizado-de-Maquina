```python
import spacy
import json
from spacy import displacy
from IPython.display import display, HTML

nlp = spacy.load("pt_core_news_lg")
with open("dados/resultados.json", "r", encoding="utf-8") as f:
    lista_de_dados = json.load(f)

options= {
    "bg": "#ebebdf", "color": "#191265", "font": "Times New Roman",
    "compact": True, "distance": 90, "arrow_width": 8, "arrow_stroke": 2,
    "word-spacing": 45, "collapse_phrases": False
}

i=1
for item in lista_de_dados[15:16]:
        print("="*50)
        print("\nTEXTO " + str(i) +":\n")
        i+=1
        texto_para_analise = item["conteudo_texto"]
        doc = nlp(texto_para_analise)
        sentence_spans = list(doc.sents)
        html = displacy.render(sentence_spans, style="dep", options=options, jupyter=False)
        display(HTML(html))


```
